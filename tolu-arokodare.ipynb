```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
player_url = "https://fbref.com/en/players/a42760bc/matchlogs/2023-2024/c37/Tolu-Arokodare-Match-Logs"
```

```python
data = requests.get(player_url)
```

```python
player_stats =  pd.read_html(data.text, match="2023-2024 Belgian Pro League Match Logs")[0]
```

```python
player_stats.head()
```

```python
player_stats.columns = player_stats.columns.droplevel()
```

```python
player_stats.to_csv("tolu.csv")
```
